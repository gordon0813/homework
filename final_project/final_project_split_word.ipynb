```python
from pyhanlp import *
def change_to_txt(a):
    savelist=[]
    for word in a.iterator():  # 通过dir()可以查看sentence的方法
        savelist.append("%s(*)%s(*)%s(*)%s(*)%s(*)%s(*)%s" % (word.LEMMA,word.ID,word.CPOSTAG
                                                  , word.DEPREL,word.HEAD.LEMMA, word.HEAD.ID
                                                       ,word.HEAD.CPOSTAG))
    return "\n>->".join(savelist)
def decode_txt(a):
    alist=a.split("\n>->")
    alist=[x.split("(*)") for x in alist]
    return alist
```

```python
txtlist=[]
l=1#有幾篇,從零開始
name="a"
for i in range(1,l+1):  
    with open("./data/article_in/"+str(name)+" ("+str(i)+").txt","r",encoding="utf8") as file:
        s1=file.read()
        s1=s1.replace("\n","")
        txtlist.append(s1)
```

```python
parser = JClass('com.hankcs.hanlp.dependency.nnparser.NeuralNetworkDependencyParser')()
parser.enableDeprelTranslator(False)
analyer = PerceptronLexicalAnalyzer()
parser.setSegment(analyer._proxy)
CustomDictionary = JClass("com.hankcs.hanlp.dictionary.CustomDictionary") 
with open("./data/newword.txt","r") as file:
    temps=file.read()
    addlist=temps.split("\n")
#addlist=["道敦堀 nsf"]#自定新词
for i in addlist:    
    CustomDictionary.add(i) 
```
